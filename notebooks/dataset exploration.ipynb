# Data cleaning

```python
#merging similar genres to the one general
#conert all value's lettercases to lower
df['Top Genre'] = (df["Top Genre"].str.strip()).str.lower()
```

```python
#convert negative values in "Loudness " column to positive by using abs() function
df['Loudness (dB)'] = df['Loudness (dB)'].abs()
```

# Data representation

### Correlation matrix

```python
plt.figure(figsize=(10,8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
```

### Attributes distribution

```python
#draw a pieplot for result
labels = df['Top Genre'].value_counts().index
sizes = df['Top Genre'].value_counts().values
plt.figure(figsize = (10,10))
plt.pie(sizes, labels=labels)
autopct=('%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
#draw a pieplot for result
labels = df['Artist'].value_counts().index
sizes = df['Artist'].value_counts().values
plt.figure(figsize = (10,10))
plt.pie(sizes, labels=labels)
autopct=('%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
#draw a pieplot for result
labels = df['Year'].value_counts().index
sizes = df['Year'].value_counts().values
plt.figure(figsize = (10,10))
plt.pie(sizes, labels=labels)
autopct=('%1.1f%%')
plt.axis('equal')
plt.show()
```
